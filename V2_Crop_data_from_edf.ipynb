```python
for session_n in edf_struct.keys():

    filename = edf_struct[session_n]['filename']
    raw = mne.io.read_raw_edf(filename, preload=False)


    crop_tmin = edf_struct[session_n]['start'] / raw.info['sfreq'] 
    crop_tmax = edf_struct[session_n]['end'] / raw.info['sfreq']
    raw = raw.crop(tmin=crop_tmin, tmax=crop_tmax)


    save_file_name = 'data/sub_'+str(edf_struct[session_n]['subject'])+'_session_'+str(edf_struct[session_n]['session'])+'_raw.fif'
    raw.save(save_file_name)
    
    
    
```

```python
ttl_info_dict = {}
for session_n in edf_struct.keys():
    ttl_info_dict[session_n] = {}
    ttl_info_dict[session_n]['ttl'] = edf_struct[session_n]['ttl']
    ttl_info_dict[session_n]['subject'] = edf_struct[session_n]['subject']
    ttl_info_dict[session_n]['session'] = edf_struct[session_n]['session']
pickle.dump( ttl_info_dict, open( "ttl_info_dict.p", "wb" ) )
```
