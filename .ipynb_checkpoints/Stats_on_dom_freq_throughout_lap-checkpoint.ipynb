```python
log_welch = np.load('../Results/log_welch_throughout_lap.npy')
```

```python
sub = 1

data = {}

for uu in uni.astype(int):

    data[uu] = []

    for tt in range(len(log_welch[sub])):

        if int(log_welch[sub][tt][3]) == uu:
            
            if uu != 13:
                data[uu].append( log_welch[sub][tt][4] )
                
                
            ## the lenght of the last trials are not consistend due to the lack of ttl in the last trial as 
            ## computed by Compute_dominant_frequency_throughout_lap. So we are ignoring the low resolution trials here. 
            if uu == 13 and len(log_welch[sub][tt][4]) == 11:   
                
                data[uu].append( log_welch[sub][tt][4] )
            
```

```python
for kk in np.sort(data.keys()):    
    plt.plot(np.mean(np.array(data[kk]), axis=0))
    plt.show()    
```
